```python
# Check for null values
null_values = housing_all.isnull().sum()
print("Null values:")
print(null_values)

# Check for zero values
zero_values = (housing_all == 0).sum()
print("\nZero values:")
print(zero_values)
```

```python
housing_all['town'] = housing_all['town'].replace('/', '_', regex=True)
print(housing_all['town'].unique())
```

```python
housing_all['flat_type'] = housing_all['flat_type'].replace('MULTI GENERATION','MULTI-GENERATION',regex = True)
print(housing_all['flat_type'].unique())
```

```python

# Loop over unique towns and create a separate CSV for each
for town in housing_all['town'].unique():
    # Initialize an empty list to collect rows
    town_rows = []

    # Iterate through each row in the DataFrame
    for _, row in housing_all.iterrows():
        if row['town'] == town:
            # Add the row to the list
            town_rows.append(row)

    # Convert the list of rows to a DataFrame and save to CSV
    town_df = pd.DataFrame(town_rows)
    town_df.to_csv(f'Town/{town}.csv', index=False)
```

```python
for flat_type_value in housing_all['flat_type'].unique():
    # Initialize an empty list to collect rows
    rows = []

    # Iterate through each row in the DataFrame
    for _, row in housing_all.iterrows():
        if row['flat_type'] == flat_type_value:
            # Add the row to the list
            rows.append(row)

    # Convert the list of rows to a DataFrame
    flat_df = pd.DataFrame(rows)


    # Save the DataFrame to a CSV file with the flat_type value as the filename
    flat_df.to_csv(f'Flat_type/{flat_type_value}.csv', index=False)
```
